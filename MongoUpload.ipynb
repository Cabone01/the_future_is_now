```python
import pandas as pd
import pymongo
from pymongo import MongoClient
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
db = client.movies_db
collection = db.scifi
db.scifi.delete_many({})
```

```python
top_grossing_movies = pd.read_csv("RawData/box_office_mojo_movies.csv")
top_grossing_movies["title"] = top_grossing_movies["title"].astype("string")
top_grossing_movies
```

```python
scifi_movies = pd.read_csv("RawData/themoviedb_api_pull_scifi.csv")
scifi_movies["title"] = scifi_movies["title"].astype("string")
scifi_movies
```

```python
#top_scifi_movies = pd.concat([scifi_movies, top_grossing_movies])
#top_scifi_movies
```

```python
top_scifi_movies = pd.merge(top_grossing_movies, scifi_movies, on = "title")
top_scifi_movies
```

```python
top_scifi_movies = top_scifi_movies.drop_duplicates()
top_scifi_movies
```

```python
for x in top_scifi_movies["title"]:
    print(x)
```

```python
top_scifi_movies = top_scifi_movies.reset_index(drop = True)
data_dict = top_scifi_movies.to_dict("records")
data_dict
```

```python
collection.insert_many(data_dict)
```
